```python
cluster_features = [
    "fatal_crash_ratio",
    "fatalities_per_10k_vehicles",
    "road_crash_fatalities",
    "road_crashes"
]
```

```python
df["fatal_crash_ratio"] = (
    df["fatal_road_crashes"] / df["road_crashes"]
)
```

```python
df["fatalities_per_10k_vehicles"] = (
    df["road_crash_fatalities"] / df["total_motor_vehicles"]
) * 10000
```

```python
df[["fatal_crash_ratio", "fatalities_per_10k_vehicles"]].head()
```

```python
cluster_features = [
    "fatal_crash_ratio",
    "fatalities_per_10k_vehicles",
    "road_crash_fatalities",
    "road_crashes"
]
```

```python
X = df[cluster_features].fillna(0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
df["cluster"] = kmeans.fit_predict(X_scaled)
```

```python
df["cluster"].value_counts()
```

```python
cluster_labels = {
    0: "High Risk",
    1: "Moderate Risk",
    2: "Lower Risk"
}

df["risk_category"] = df["cluster"].map(cluster_labels)
```

```python
df.to_csv(
    "../data/cleaned/master_yearly_with_clusters.csv",
    index=False
)
```
