```python
def clean_columns(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
        .str.replace("/", "_")
        .str.replace("-", "_")
    )
    return df
```

```python
road_df = pd.read_csv("../data/raw/road_crash_trends_2014_2023.csv")
```

```python
road_df = clean_columns(road_df)

road_df["year"] = road_df["year"].astype(int)

numeric_cols = [
    "road_crashes",
    "road_crash_fatalities",
    "fatal_road_crashes"
]
road_df[numeric_cols] = road_df[numeric_cols].astype(int)
```

```python
road_df.to_csv(
    "../data/cleaned/road_crash_trends_cleaned.csv",
    index=False
)
```

```python
vehicle_totals = pd.read_csv("../data/raw/vehicle_totals_2014_2023.csv")
```

```python
vehicle_totals = clean_columns(vehicle_totals)
vehicle_totals["year"] = vehicle_totals["year"].astype(int)

for col in vehicle_totals.columns:
    if col != "year":
        vehicle_totals[col] = vehicle_totals[col].astype(int)
```

```python
vehicle_totals.to_csv(
    "../data/cleaned/vehicle_totals_cleaned.csv",
    index=False
)
```

```python
vehicle_growth = pd.read_csv("../data/raw/vehicle_growth_2014_2023.csv")
```

```python
vehicle_growth = clean_columns(vehicle_growth)
vehicle_growth["year"] = vehicle_growth["year"].astype(int)

for col in vehicle_growth.columns:
    if col not in ["year", "yearly_growth_percent"]:
        vehicle_growth[col] = vehicle_growth[col].astype(int)

vehicle_growth["yearly_growth_percent"] = vehicle_growth[
    "yearly_growth_percent"
].astype(float)
```

```python
vehicle_growth.to_csv(
    "../data/cleaned/vehicle_growth_cleaned.csv",
    index=False
)
```

```python
victim_df = pd.read_csv("../data/raw/victim_category_2019_2023.csv")
```

```python
victim_df = clean_columns(victim_df)
victim_df["year"] = victim_df["year"].astype(int)

for col in victim_df.columns:
    if col != "year":
        victim_df[col] = victim_df[col].astype(int)
```

```python
victim_df.to_csv(
    "../data/cleaned/victim_category_cleaned.csv",
    index=False
)
```

```python
vehicle_type_df = pd.read_csv(
    "../data/raw/vehicle_type_crashes_2021_2023.csv"
)
```

```python
vehicle_type_df = clean_columns(vehicle_type_df)

vehicle_type_df["vehicle_type"] = (
    vehicle_type_df["vehicle_type"]
    .str.strip()
    .str.lower()
)
```

```python
vehicle_type_df.to_csv(
    "../data/cleaned/vehicle_type_crashes_cleaned.csv",
    index=False
)
```
